### Montado

En el siguiente script se muestra como realizar un montado a Databricks

```
configs = {"fs.azure.account.auth.type": "OAuth",
           "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
           "fs.azure.account.oauth2.client.id": dbutils.secrets.get(scope = "keybi64pro", key = "loggin"),
           "fs.azure.account.oauth2.client.secret" : dbutils.secrets.get(scope = "keybi64pro", key = "password"),
           "fs.azure.account.oauth2.client.endpoint": "https://login.microsoftonline.com/e7bcc36a-338a-4c34-9418-60f44d2712d6/oauth2/token"}
dbutils.fs.mount(
  source = "abfss://aitour@storagebi64pro.dfs.core.windows.net/",
  mount_point = "/mnt/aitour",
  extra_configs = configs)
```

### Carga de Datos

En el siguiente script se muestra como cargar csvs a Databricks

```
productosDF = spark.read.format('csv').options(header='true', inferschema='true').load("/mnt/aitour/Productos1.csv")
productosDF.write.format("parquet").saveAsTable("productos")
```

```
%sql
SELECT * FROM productos
```

### Desmontado

En el siguiente script se muestra como desmontar de Databricks

```
dbutils.fs.unmount("/mnt/aitour")
```
