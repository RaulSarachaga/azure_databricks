### Carga de Datos

En el siguiente script se muestra como cargar csvs a Databricks

```
ventasDF = spark.read.format('csv').options(header='true', inferschema='true').load("/mnt/aitour/Ventas.csv")
ventasDF.write.format("parquet").saveAsTable("ventas")
```

```
clientesDF = spark.read.format('csv').options(header='true', inferschema='true').load("/mnt/aitour/Clientes.csv")
clientesDF.write.format("parquet").saveAsTable("clientes")
```

```
productosDF = spark.read.format('csv').options(header='true', inferschema='true').load("/mnt/aitour/Productos1.csv")
productosDF.write.format("parquet").saveAsTable("productos")
```

```
calendarioDF = spark.read.format('csv').options(header='true', inferschema='true').load("/mnt/aitour/Calendario.csv")
calendarioDF.write.format("parquet").saveAsTable("calendario")
```
